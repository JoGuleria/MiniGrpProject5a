```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1 = nx.read_edgelist("C:\\Users\\jyoti\\Simon MSBA\\Spring 2025\\Spring B\\Social Media and Text analytics\\S11 Identify Influencers\\graph1.edgelist",nodetype=int)
```

```python
degree_centrality = nx.degree_centrality(G1)
eignvector_centrality = nx.eigenvector_centrality(G1)
betweenness_centrality = nx.betweenness_centrality(G1)
closeness_centrality = nx.closeness_centrality(G1)
clustering_coefficient = nx.clustering(G1)
```

```python
print(degree_centrality)
print(eignvector_centrality)
print(betweenness_centrality)
print(closeness_centrality)
print(clustering_coefficient)
```

```python
#nodes with the highest degree of centrality 
max_degree_node = max(degree_centrality, key = degree_centrality.get)
max_degree_value = degree_centrality[max_degree_node]
print(f"Node with the highest degree centraility: {max_degree_node}(Degree centrality : {max_degree_value:4f})")
```

```python
# Top 5 candidates
top_5_nodes = sorted(degree_centrality.items(), key=lambda x:x[1], reverse =True)[:5]
print('Top 5 Candidates:')
for node, value in top_5_nodes:
    print(f'Node{node}:{value:.4f}')
```

```python
# visualisation network
node_sizes = [v*2000 for v in degree_centrality.values()]
plt.figure(figsize = (10,8) )
pos = nx.spring_layout(G1, seed = 42)
nx.draw(G1,pos, with_labels= True, node_color = 'lightblue', edge_color = 'gray', node_size = node_sizes)
plt.title("Social Network with Node Sizes Proportional To Degree Centrality")
plt.show
```
